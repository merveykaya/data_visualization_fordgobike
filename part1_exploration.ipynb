## Introduction

According to [Wiki/BayWheels](https://en.wikipedia.org/wiki/Bay_Wheels), Bay Wheels is the first regional public bicycle sharing system in the U.S. that services more than just a single city or adjacent cities.On June 28, 2017, the system officially re-launched as Ford GoBike.
Coinciding with the expansion and rebranding as Ford GoBike, several new programs such as integration with Clipper cards and a low-income pricing option were launched. Ford GoBike expanded electric bicycles by 250. 
In 2018, the Bay Wheels system reached around 10,000 annual subscribers, over 2,600 bicycles in 262 stations across San Francisco, East Bay and San Jose. However, in 2019 the system rebranded to Bay Wheels and dropped the Ford naming, with funding motivation of another company.

## Preliminary Wrangling

The Ford GoBike data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area. Each rows in the data represents a user's trip.

In this analysis, I would like to see how Ford GoBike expanded this system, and will answer these questions:

* When are most trips taken in terms of time of day and day of the week?

* How long does the average trip take?

* Does the above depend on if a user is a subscriber or customer?

* Does the above depend on user's gender?

* What age group rides bikes the most?

* What is the relation between these variables?

## Exploration


```python
df_ford = df.copy()
df_ford.info()
```

I added new columns for duration duration_day, duration_hour and duration_minutes.

```python
df_ford['duration_day'] = pd.to_datetime(df_ford['start_time']).dt.day_name()
```

```python
df_ford['duration_hour'] = pd.to_datetime(df_ford['start_time']).dt.hour
```

```python
df_ford['duration_minutes'] = df_ford['duration_sec'] // 60
```

I added a column for member age, base year selected as 2019.

```python
df_ford['member_age'] = 2019 - df_ford['member_birth_year']
```

There are some irrelavent birth years in the data, I assumed that someone older than 80 years old cannot ride a bike. 

```python
df_ford.drop(df_ford[df_ford.member_birth_year < 1939].index, inplace=True)
df_ford.member_birth_year.min()
```

Test for data coverage:

```python
data_coverage = pd.DataFrame(((df_ford.count()-df_ford.isnull().sum())/df_ford.count())*100 , columns=['Data Coverage'])
data_coverage
```

```python
ax = data_coverage.plot(kind = 'bar' , figsize = (10,9))
ax.set(xlabel = 'Variables' , ylabel = 'Data Coverage')

plt.show()
```

The data looks normal, let's look for the null values.

```python
df_ford.isnull().sum()
```

```python
na_counts = df_ford.drop(['start_station_id', 'start_station_name', 'end_station_id', 'end_station_name', 'member_birth_year', 'member_gender', 'member_age'], axis=1).isna().sum()
na_counts
```

## Univariate Exploration

```python
# nominal data : ordered by frequency of data points
gender_counts = df_ford['member_gender'].value_counts()
gender_order = gender_counts.index
gender_order
```

```python
#Proportion of max count in gender = Male proportion
n_df_ford = df_ford.shape[0]
male_counts = gender_counts[0]
male_prop = male_counts / n_df_ford
female_counts = gender_counts[1]
female_prop = female_counts / n_df_ford
print('male proportion is:' , male_prop)
print('female proportion is:' , female_prop)
```

```python
tick_props = np.arange(0, male_prop, 0.1)
tick_names = ['{:0.2f}'.format(v) for v in tick_props]
tick_names
```

```python
sb.countplot(data = df_ford, x='member_gender',order=gender_order);
plt.yticks(tick_props * n_df_ford, tick_names)
plt.ylabel('proportion');
```

Male proportion is about 71% while female proportion is approximately 22%.
The plot shows that males are using the system more frequently.

```python
#user_type
user_type_counts = df_ford['user_type'].value_counts()
user_type_order = user_type_counts.index
user_type_order
```

## Bivariate Exploration



```python
sb.regplot(data = df_ford, x='duration_hour', y='duration_minutes');
plt.xlabel('Hour of duration')
plt.ylabel('Length of duration(min)');
```

The plot suggests no correlation between duration length and duration hour. Two peaks I observed on duration hours are significant for the usage duration of bikes. However, it can be said that people prefer to ride in longer durations after 13:00.

```python
sb.stripplot(x=df_ford["user_type"] , y=df_ford['duration_sec'] , jitter=True , alpha = 0.25)
plt.show()
```

From this plot, we can see that there is a non-monotanic relationship between customer type and duration length. Subscriber tends to take shorter trips than customers. 

```python
sb.boxplot(data= df_ford, x='member_gender', y='duration_hour');
plt.xticks(rotation = 15);
```

There is no relation between starting hour of duration and gender.

```python
sb.boxplot(data= df_ford, x='user_type', y='duration_hour');
plt.xticks(rotation = 15);
```

First quartile(minimum) starting hour of duration for subscribers is earlier then customers. Their medians and maxs are approximately equal. We may say that subscribers rent in a more wide of range in a day, as starting to rent earlier. It is probably the time for going to office.

## Multivariate Exploration

```python
m = sb.FacetGrid(data=df_ford ,col = 'user_type' , hue = 'member_gender')
m.map(plt.scatter , 'user_type','duration_minutes' , alpha = 0.5).add_legend()
m.set_xlabels('User Type')
m.set_ylabels('Duration Length(minutes)')
m.set_xticklabels()
plt.show()
```

The scatter plot shows that female tends to have longer minutes biking than males. Also, subscriber females have longer trips than customer females. Customer females prefer to have less longer trips than customer males.

```python
m = sb.FacetGrid(data=df_ford ,col = 'member_gender' , hue = 'user_type')
m.map(plt.scatter , 'member_age','duration_hour' , alpha = 0.5).add_legend()
m.set_xlabels('Age')
m.set_ylabels('Duration hour')
m.set_xticklabels()
plt.show()
```

```python
a = sb.FacetGrid(data=df_ford ,col = 'member_gender' , hue = 'user_type')
a.map(plt.scatter , 'member_age','duration_sec' , alpha=0.5).add_legend()
a.set_xlabels('Age')
a.set_ylabels('Duration in seconds')
a.set_xticklabels()
plt.show()
```

The average of subscribers' ages are similar about durations for both duration hour and duration length on gender in case they are customers or subscribers.

```python
duration_day_counts = df_ford['duration_day'].value_counts()
duration_day_order = duration_day_counts.index
duration_day_order
```

```python
d = sb.factorplot(x="user_type", y="duration_minutes", col = "duration_day",hue="member_gender",col_wrap=4, kind="box",
                  data=df_ford.query('duration_minutes < 55'))
d.set_xlabels('User type')
d.set_ylabels('Duration Length(minutes)')
plt.show()
```

To see the comparisons more clear, I set the duration length as less than 60 minutes because I already know people tend to ride bikes less than an hour. Previously, I've already found out that people prefer to rent a bike mostly on Wednesday. 

Also, from the week-day comparison based on user_type and duration length represents a trend for both customer males and females are more tend to ride longer on the Weekends.
